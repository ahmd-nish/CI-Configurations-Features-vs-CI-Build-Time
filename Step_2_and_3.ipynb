```python
# shared session for pooling
session = requests.Session()
session.headers.update(headers)
```

```python
def check_and_append(row):
    owner, repo = row["name"].split("/", 1)
    url = f"https://api.github.com/repos/{owner}/{repo}/contents/.github/workflows"
    try:
        r = session.get(url, timeout=10)
        time.sleep(sleep_between)
        if r.status_code == 200:
            pd.DataFrame([row]).to_csv(out_csv, mode="a", header=False, index=False)
            logger.info(f"{owner}/{repo}: workflows found")
            return True
        elif r.status_code != 404:
            logger.warning(f"{owner}/{repo}: HTTP {r.status_code}")
    except Exception as e:
        logger.error(f"{owner}/{repo}: {e}", exc_info=True)
    return False
```

```python
found = 0
with ThreadPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(check_and_append, row) for _, row in df.iterrows()]
    for i, fut in enumerate(as_completed(futures), start=1):
        if fut.result():
            found += 1
        if i % 100 == 0 or i == total:
            print(f"[{i}/{total}] checked, {found} workflows found")

logger.info(f"Step 3 complete: {found}/{total} repos use GitHub Actions")
print(f"Done Step 3: {found}/{total} repos use GitHub Actions")
```

```python
import os
import time
import logging
from pathlib import Path

import pandas as pd
import requests

# ─── CONFIG ────────────────────────────────────────────────────────────────────
IN_CSV        = Path("projects_active.csv")
OUT_CSV       = Path("projects_ci.csv")
LOG_FILE      = Path("ci_graphql.log")
BATCH_SIZE    = 20            # number of repos per GraphQL query
SLEEP_BETWEEN = 1             # seconds between batches
# ────────────────────────────────────────────────────────────────────────────────
```

```python
# 1) Load & authenticate
token = os.getenv("GITHUB_TOKEN") or input("GitHub token: ").strip()
if not token:
    raise RuntimeError("GitHub token is required")

headers = {
    "Authorization": f"bearer {token}",
    "Accept":        "application/vnd.github.v3+json"
}
```

```python
# Quick check
resp = requests.post("https://api.github.com/graphql",
                     json={"query": "{ viewer { login }}"},
                     headers=headers)
resp.raise_for_status()
print("Authenticated as", resp.json()["data"]["viewer"]["login"])
```

```python
# 2) Setup logging
logger = logging.getLogger("ci_graphql")
logger.setLevel(logging.INFO)
fh = logging.FileHandler(LOG_FILE, mode="w")
fh.setFormatter(logging.Formatter("%(asctime)s %(levelname)s %(message)s"))
logger.addHandler(fh)
logger.addHandler(logging.StreamHandler())
```

```python
# 3) Read input and prepare output
if not IN_CSV.exists():
    raise FileNotFoundError(f"{IN_CSV} not found")

df = pd.read_csv(IN_CSV)
total = len(df)
logger.info(f"Loaded {total} repos to check")

# Write header to output CSV
df.head(0).to_csv(OUT_CSV, index=False)

# Helper to build a single GraphQL query for one batch
def make_query(batch):
    parts = []
    for i, row in batch.iterrows():
        owner, repo = row["name"].split("/", 1)
        alias = f"r{i}"
        parts.append(f'''
  {alias}: repository(owner:"{owner}", name:"{repo}") {{
    wf: object(expression:"HEAD:.github/workflows") {{
      __typename
    }}
  }}''')
    return "{\n" + "\n".join(parts) + "\n}"
```

This GraphQL‐based check batches dozens of repos into a single request, then inspects each aliased result for a `.github/workflows` “Tree” entry, writing survivors incrementally and respecting rate limits.

- **Batch+Alias:** Build one GraphQL query per batch (e.g. 20 repos) using aliases (`r0`, `r1`, …) and check each `data[alias].wf.__typename == "Tree"`.  
- **Incremental & Resilient:** Append each positive match to the CSV as it arrives, catching/logging errors and backing off on failures.  
- **Throttled Progress:** Sleep briefly between batches and log “checked X/Y, found Z” to avoid rate‐limit hits and monitor progress.  

```python
# 4) Batch‐wise GraphQL fetch
survivors = []
for start in range(0, total, BATCH_SIZE):
    batch = df.iloc[start : start + BATCH_SIZE]
    query = make_query(batch)
    try:
        r = requests.post("https://api.github.com/graphql",
                          json={"query": query},
                          headers=headers,
                          timeout=30)
        r.raise_for_status()
        data = r.json().get("data", {})
    except Exception as e:
        logger.error(f"GraphQL batch {start}-{start+BATCH_SIZE} failed: {e}")
        time.sleep(SLEEP_BETWEEN * 2)
        continue

    # 5) Inspect each alias for existence of a Tree (workflows folder)
    for i, row in batch.iterrows():
        alias = f"r{i}"
        entry = data.get(alias)
        if entry and entry.get("wf") and entry["wf"]["__typename"] == "Tree":
            pd.DataFrame([row]).to_csv(OUT_CSV, mode="a", header=False, index=False)
            survivors.append(row["name"])
            logger.info(f"{row['name']}: workflows found")

    # Progress report
    checked = min(start + BATCH_SIZE, total)
    logger.info(f"Checked {checked}/{total} repos, found {len(survivors)} survivors so far")
    time.sleep(SLEEP_BETWEEN)

logger.info(f"Step 3 complete: {len(survivors)}/{total} repos use GitHub Actions")
print(f"Done Step 3: {len(survivors)}/{total} repos use GitHub Actions")
```
